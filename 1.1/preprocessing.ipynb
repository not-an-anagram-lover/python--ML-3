```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\\Users\\lenovo\\PycharmProjects\\Music-genre-classification\\tracks.csv",na_values='[]',low_memory=False) 
tracks = df.iloc[2:,[0,40]]
tracks=tracks.rename({'Unnamed: 0':'Track_ID', 'track.7':'genre_top', 'track.9':'genres_all'}, axis=1)

tracks=tracks.set_index("Track_ID")
tracks.head(20)
```

```python
echonest=pd.read_csv("C:\\Users\\lenovo\\PycharmProjects\\Music-genre-classification\\echonest.csv",low_memory = False,index_col=0)
echonest_audio_features = echonest.iloc[3:,[0,1,2,3,4,5,6,7]]
echonest_audio_features.index.rename("Track_ID",inplace=True)
echonest_audio_features=echonest_audio_features.rename({'echonest':'acousticness', 'echonest.1':'danceability', 'echonest.2':'energy', 'echonest.3':'instrumentalness', 'echonest.4':'liveness', 'echonest.5':'speechiness', 'echonest.6':'tempo', 'echonest.7':'valence'}, axis=1)
echonest_audio_features.head(20)
echonest_audio_features.to_csv("C:\\Users\\lenovo\\PycharmProjects\\Music-genre-classification\\echonest_audio_features.csv")
```

```python

data_with_target = pd.merge(echonest_audio_features,tracks,how = "left",left_index = True , right_index = True )

data_with_target.rename({'genre_top':'Target_Genre'},axis=1,inplace=True)
data_with_target.head(20)
```

```python
#removing missing values from target variable
data_with_target.dropna(axis=0,how='any', inplace = True)
data_with_target.head(20)
```

```python
#tokenizing Target

print("Unique genres before tokenizing : \n" ,np.unique(list(data_with_target.Target_Genre)))
```

```python
#print(data_with_target.loc[data_with_target['Target_Genre']==-1])
def nullColumns(dataset):
    null_columns = []
    for column in dataset.columns.tolist():
        if(dataset[column].isnull().sum() > 0):
            print(column)
            null_columns = null_columns.append(column)
nullColumns(data_with_target)
def dropNull(dataset, null_list):
    dataset.dropna(axis=0,how='any', inplace = True)
    return dataset

data_with_target = dropNull(data_with_target, nullColumns)
print('The shape of  target variable after dropping NAN is', data_with_target.shape)
print('\nThe number of unique values in target variable after dropping NAN is', data_with_target.nunique())
print('\n\nThe value counts of each target variable after dropping NAN is\n', data_with_target.Target_Genre.value_counts())
print("Unique genres before tokenizing : \n" ,np.unique(list(data_with_target.Target_Genre)))
obj_df = data_with_target.select_dtypes(include=['object']).copy()
obj_df["Target_Genre"] = obj_df["Target_Genre"].astype('category')
obj_df["Target_Genre"] = obj_df["Target_Genre"].cat.codes

print("Uniques : \n" ,np.unique(list(obj_df.Target_Genre)))
obj_df.head(20)
```

```python
obj_df.to_csv("C:\\Users\\lenovo\\PycharmProjects\\Music-genre-classification\\ready_data.csv")
```
